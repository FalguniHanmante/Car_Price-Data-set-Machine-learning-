### Gride Search CV

```python
from sklearn.feature_selection import RFE
# step-1: create a cross-validation scheme
folds = KFold(n_splits = 10, shuffle = True, random_state = 100)

# step-2: specify range of hyperparameters to tune
hyper_params = [{'n_features_to_select': list(range(1, 14))}]


# step-3: perform grid search
# 3.1 specify model
lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm)      
```

```python
from sklearn.model_selection import GridSearchCV
model_cv = GridSearchCV(estimator = rfe, 
                        param_grid = hyper_params, 
                        scoring= 'r2', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
model_cv.fit(X_train, y_train) 
```

```python
model_cv.best_estimator_
```

```python
# final model
n_features_optimal = 2

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, n_features_to_select=n_features_optimal)             
rfe = rfe.fit(X_train, y_train)

# predict prices of X_test
y_pred = lm.predict(X_test)
import sklearn.metrics
r2 = sklearn.metrics.r2_score(y_test, y_pred)
print(r2)
```

### Handling Outlier

```python
plt.figure(figsize =(8,4))
sns.boxplot(x = 'enginesize', data=car_1, orient='h')
```

```python
from scipy import stats

R_q1 = car_1['enginesize'].quantile(q=0.25)
R_q3 = car_1['enginesize'].quantile(q=0.75)
print('1st quartile = {} \n3rd quartile range = {} \ninter quartile range = {} '.format(R_q1,R_q3,stats.iqr(car_1['enginesize'])))
```

```python
lower_outlier = R_q1-1.5*(R_q3-R_q1)
upper_outlier = R_q3+1.5*(R_q3-R_q1)

print('lower outlier = ',lower_outlier)
print('uppper outlier = ',upper_outlier)
```

```python
print('Number of upper outlier =  ',car_1[car_1['enginesize']>400]['enginesize'].count())
print('Number of lower outlier =  ',car_1[car_1['enginesize']<31]['enginesize'].count())
```

```python
for cols in car_1.columns[:-1]:
    Q1 = car_1[cols].quantile(q=0.25)
    Q3 = car_1[cols].quantile(q=0.75)
    iqr=Q3-Q1
    
    low = Q1-1.5*(Q3-Q1)
    high = Q3+1.5*(Q3-Q1)
    car_1.loc[(car_1[cols]<low) | (car_1[cols]>high),cols] = car_1[cols].median()
```

```python
plt.figure(figsize =(8,4))
sns.boxplot(x = 'enginesize', data=car_1, orient='h')
```
